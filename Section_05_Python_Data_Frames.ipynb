```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
#Distribution
vis1 = sns.distplot(stats['Internetusers'], bins = 30) # recall that shift + tab will open up a reference section about
                                                        #the functon
```

```python
#Boxplots:
vis2 = sns.boxplot(data = stats, x = "IncomeGroup", y = "Birthrate")
```

```python
#Seaborn gallery https://seaborn.pydata.org/examples/index.html
```

```python
#Initially there was a line of best fit because fit_reg = True is default. We changed it to False because visually
# the data is very spread from the line of best fit. Their appears to be a negative relation ship 
# 
```

```python
vis3 = sns.lmplot(data=stats, x= 'Internetusers', y ='Birthrate',\
                  fit_reg = False, hue = 'IncomeGroup' )  #linear model plot
```

```python
# a facet grids take up a lot of space and have their own parameter to change their size.
```

```python
##Keyword arguments in Python
```

```python
#marker size: when you hit shit + tab to see how to change marker size we see there is not explicit parameter.
# so there is a parameter we want to lookup in seaborn.lmplot. 
# {scatter,line}_kws : dictionaries
    #Additional keyword arguments to pass to ``plt.scatter`` and
    #``plt.plot``. 
    # https://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot
#Seaborn relies on matplotlib. 
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter
# So there are other parameters in pyplot that are not in the parameter list of the seaborn function.
```

```python
vis3 = sns.lmplot(data=stats, x= 'Internetusers', y ='Birthrate',\
                  fit_reg = False, hue = 'IncomeGroup', \
                  scatter_kws={'s':100}) 
```
