The sample mean of the sample distribution converges to the expected value when the sample size, n, goes to infinity.

Test law of large numbers for N random normally distributed numbers s.t. mean = 0 and stdev = 1

Create Python script that counts how many of these numbers fall between -1 and 1 and divide by the total quantity of N

E(X) = 68.2%

Check that the Mean of (X_n) -> (EX) as you rerun your script with increasing N (where N is the sample size)

```python
import numpy as np
from numpy.random import randn
N = 100000   #specify sample size
counter = 0 #reset counter
for i in randn(N): #iterate over random values
    if -1 < i < 1: #check where iterated variable falls in terms of E(X) = 68.2
        counter = counter + 1 #increase counter if condition is met
answer = counter / N #calculate hit-ratio
print(answer)    #print answer
```
