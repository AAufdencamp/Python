```python
list1 = []
mylabels = []
for i in movies.Genre.cat.categories:
   list1.append(movies[movies.Genre == i].BudgetMillions)
   mylabels.append(i)
    
sns.set_style("whitegrid")    
fig, ax = plt.subplots() #we are using plt.subplot to establish one plot as a hack so we can control the plot size of h.
fig.set_size_inches(11.7, 8.27) #size of A4 paper
h = plt.hist(list1, bins = 30, stacked = True, rwidth = 1, label = mylabels)
plt.title("Movie Budget Distribution", fontsize = 35, color = 'DarkBlue', fontname = "Arial")
plt.ylabel("Number of Movies", fontsize = 25, color='Red')
plt.xlabel("Budget", fontsize = 25, color = 'Green')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(frameon=True, fancybox=True, shadow = True, framealpha =1, prop ={'size':20}) #to show the labels we just created
plt.show()
```
