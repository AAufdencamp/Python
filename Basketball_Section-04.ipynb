```python
#Fix up the inputs.Want the plot to be able to accept a matrix as input
#now our functions requires to input variables: the matrix (ie data) we care about and the playerlist
#lastly we set our function argument playerlist = Players, which is a pre-established list of players, which serves
#as default value if none is passed into the function when calling it.

def myplot(data, playerlist = Players):
    Col = {"KobeBryant":'Black',"JoeJohnson":'Red',"LeBronJames":'Green',"CarmeloAnthony":'Blue',"DwightHoward":'Magenta',"ChrisBosh":'Black',"ChrisPaul":'Red',"KevinDurant":'Green',"DerrickRose":'Blue',"DwayneWade":'Magenta'}
    Marker = {"KobeBryant":'s',"JoeJohnson":'o',"LeBronJames":'^',"CarmeloAnthony":'D',"DwightHoward":'s',"ChrisBosh":'o',"ChrisPaul":'^',"KevinDurant":'D',"DerrickRose":'s',"DwayneWade":'o'}
    for name in playerlist:
        plt.plot(data[Pdict[name]], c = Col[name], ls='--', marker = Marker[name], ms = 7, label = name)
    plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)),Seasons, rotation = 'vertical') 
    plt.show() 
```
